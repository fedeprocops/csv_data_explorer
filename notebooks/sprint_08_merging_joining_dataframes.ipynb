```python
customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Mario", "Luca", "Anna", "Sara"]
})
```

```python
orders = pd.DataFrame({
    "order_id": [101, 102, 103],
    "customer_id": [1, 1, 3],
    "amount": [50, 20, 80]
})
```

```python
inner_join = pd.merge(customers, orders, on="customer_id", how="inner")

left_join = pd.merge(customers, orders, on="customer_id", how="left")

outer_join = pd.merge(customers, orders, on="customer_id", how="outer")
```

```python
inner_join = pd.merge(customers, orders, on="customer_id", how="inner")
```

```python
debug_merge = pd.merge(
    customers,
    orders,
    on="customer_id",
    how="outer",
    indicator=True
)

debug_merge
```

```python
validated = pd.merge(
    customers,
    orders,
    on="customer_id",
    validate="one_to_many"
)

validated
```

```python
products = pd.DataFrame({
    "product_id": [1, 2, 3],
    "product_name": ["Laptop", "Mouse", "Keyboard"]
})


sales = pd.DataFrame({
    "sale_id": [1, 2, 3, 4],
    "product_id": [1, 1, 2, 4],
    "quantity": [2, 1, 5, 3]
})
```

```python
products
```

```python
product_sales = pd.merge(
    products, sales,
    on="product_id",
    how="left",
    )

```
