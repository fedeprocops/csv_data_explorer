# SPRINT 5 - Combining Datasets (merge, join, concat)

```python
import pandas as pd
```

```python
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'customer': ['Anna', 'Marco', 'Luca', 'Giulia']
})
```

```python
purchases = pd.DataFrame({
    'customer_id': [1, 2, 5],
    'amount': [100, 50, 200]
})
```

```python
pd.merge(customers,
         purchases,
         on='customer_id',
         how='inner')
```

```python
pd.concat([customers, purchases])
```
