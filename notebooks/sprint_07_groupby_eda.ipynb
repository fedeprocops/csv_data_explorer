Funzione Apply: apply() è una funzione di Pandas che applica una stessa regola o funzione a tutti gli elementi di una colonna (Series) o di un DataFrame, permettendo di trasformare i dati o di crearne di nuovi senza modificare quelli originali.

df['category'] = df['release_year'].apply(
    lambda x: 'Recent' if x >= 2020 else 'Classic'
)

apply():

prende ogni valore della colonna release_year;
applica la stessa regola a ogni anno;
restituisce il risultato;
crea una nuova colonna chiamata category.

Quando si utilizza apply()?
* trasformare i valori di una colonna;
* creare una nuova variabile (feature);
* applicare una regola personalizzata a ogni riga;
* arricchire il dataset senza perdere i dati originali.

```python
df = pd.read_csv('netflix_titles_cleaned.csv')
```

```python
df['category'] = df['release_year'].apply(
    lambda x: 'Recent' if x >= 2020 else 'Classic'
)
```

```python
df[['release_year', 'category']].head(10)
```

apply() non può essere utilizzato per Duration (che è una stringa di tipo object)

**Problema: la colonna duration contiene valori testuali (90 min o 120 min)**
Per classificare la durata dei film non ci interessa la parola "min" ma solo il valore numerico.
Soluzione: Filtrare il dataset solo sui Movie, Estrarre il numero dalla stringa duration; convertire il numero in un valore numerico e applicare una business rule per classificare i film in short, medium e Long.

**Soluzione:** utilizziamo il comando *str.replace* che sostituisce la sottostringa 'min' con una stringa vuota (' ') per ogni valore della colonna duration. Il risultato è una colonna contenente solo il valore numerico espresso come testo.

```python
movies = df[df['type'] == 'Movie']
```

```python
movies[['type', 'duration']].head(10)
```

```python
movies['duration'].str.replace(' min', '')
```

Funzione *astype()*: è una funzione di Pandas che permette di convertire il tipo di dato di una colonna in un altro tipo, ad esempio da stringa (object) a intero (int) o a numero decimale (float).

Pulizia del contenuto: .str.replace()
Conversione del tipo: .astype()


Classifica i film in base alla durata;
Business Rule:
Short → meno di 90 minuti
Medium → da 90 a 120 minuti (estremi inclusi)
Long → oltre 120 minuti

```python
movies['duration'].dtype
```

```python
movies['duration'].head()
```

```python
movies['duration'] = movies['duration'].str.replace(' min', '', regex=False)
```

```python
movies['duration'] = movies['duration'].astype(int)
```

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration'] = movies['duration'].str.replace(' min', '', regex=False)

movies['duration'] = movies['duration'].astype(int)
```

```python
movies['duration'].dtype
```

```python
movies['duration_category'] = movies['duration'].apply(
    lambda x: 'Short' if x <= 90
    else 'Medium' if x <= 120
    else 'Long'
)
```

```python
movies[['duration', 'duration_category']].head(10)
```

# **SPRINT 7 - Analize and resume data with ggroupby()**

```python
movies.groupby('duration_category').size()
```

```python
movies.groupby('duration_category')['duration'].mean()
```

Size() indica la frequenza; mean() indica la media quindi la frequenza con cui un evento si verifica non determina il valore medio della variabile associata.
Frequenza → Quanti film appartengono a una categoria?
Media → Quanto durano, in media, i film di quella categoria?

Pattern di groupby()= da solo non fa molto, tuttavia possiamo aggiungere altre informazioni per ottenere dei metodi:
*   .size()-> quanti elementi ci sono?
*   .count()-> quanti valori non nulli ci sono?
*   .mean()-> qual è la media?
*   .sum()-> qual è la somma?
*   .min()-> qual è il valore minimo?
*   .max()-> qual è il valore massimo?
Groupby()-> divide i dati in gruppi
Aggregate o agg()->riassume ogni gruppo con una o più statistiche.
movies.groupby('duration_category')['duration'].agg(['mean', 'min', 'max']) in forma di list

```python
movies.groupby('duration_category')['duration'].agg(['mean', 'min', 'max'])
```

Otteniamo un DataFrame perché l'aggregazione produce più colonne di risultati (mean, min e max) per ogni gruppo."

Per rinominare le colonne in maniera più incisiva e facendolo in unico passaggio faremo la funzione che segue:

```python
movies.groupby('duration_category').agg(
    durata_media=('duration', 'mean'),
    durata_minima=('duration', 'min'),
    durata_massima=('duration', 'max')
)
```
